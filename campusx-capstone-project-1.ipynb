```python
# Columns to drop -> property_name, link, property_id
df.drop(columns=['link','property_id'], inplace=True)
```

```python

df.head()
```

```python
# rename columns
df.rename(columns={'area':'price_per_sqft'},inplace=True)
df.head()
```

```python
# society
#df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
# price
#df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
def treat_price(x):
    if not isinstance(x, str):  # Check if x is not a string
        return x  # Return x as is, or handle non-string values as needed
    x = x.split(' ')  # Split the string
    if len(x) < 2:
        return x[0]  # Return the first element or handle it as needed
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        else:
            return x[0]  # Return the first element or handle other cases as needed


```

```python
# Apply the modified function
df['price'] = df['price'].apply(treat_price)
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹', '').str.replace(',', '').str.strip()
df['price_per_sqft'] = pd.to_numeric(df['price_per_sqft'], errors='coerce')
```

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0)
df['bedRoom'] = pd.to_numeric(df['bedRoom'], errors='coerce').astype('Int64')  # Using 'Int64' to handle NaN values
```

```python
# bathroom
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0)
df['bathroom'] = pd.to_numeric(df['bathroom'], errors='coerce').astype('Int64')  # Using 'Int64' to handle NaN values
```

```python
# balcony
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
# additionalRoom
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].value_counts().shape
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python

df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['floorNum'].isnull().sum()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```
